```python
df['Date'] = pd.to_datetime(df['Date']).dt.date          

#Converted Datatype of Date and removed timestamp from it as we have separate time column
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S')
```

```python
df['Hour'] = df['Time'].dt.hour
```

```python
def time_cat(x):
    if 12 >= x >= 6:
        return 'Morning'
    elif 16 >= x > 12:
        return 'Afternoon'
    elif 20 >= x > 16:
        return 'Evening'
    else:
        return 'Night'

df['Hour'] = df['Hour'].apply(time_cat)
```

```python
df.drop(columns=['Time'], inplace=True)
```

#### We have extracted hour from time and categorised it into Four parts of the day for better interpretation
